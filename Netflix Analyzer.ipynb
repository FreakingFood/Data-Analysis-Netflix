```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("Netflix Userbase.csv")
```

```python
avg_revenue_by_subscripn=data.groupby("Subscription Type")["Monthly Revenue"].mean()
```

```python
print(avg_revenue_by_subscripn)
```

```python
user_count_by_country=data.groupby("Country")["User ID"].count()
```

```python
print(user_count_by_country)
```

```python
median_age_by_plan_duration=data.groupby("Plan Duration")["Age"].median()
```

```python
print(median_age_by_plan_duration)
```

```python
total_reven_by_gender=data.groupby("Gender")["Monthly Revenue"].sum()
```

```python
print(total_reven_by_gender)
```

```python
avg_age_of_device=data.groupby("Device")["Age"].mean()
```

```python
print(avg_age_of_device)
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="Age",y="Monthly Revenue",hue="Gender",data=data)
plt.title("Age v/s Monthly revenue")
plt.xlabel("Age")
plt.ylabel("Monthly revenue")
plt.show()
```

```python
avg_revenue_by_country=data.groupby("Country")["Monthly Revenue"].mean().reset_index()
sns.barplot(x="Country",y="Monthly Revenue",data=avg_revenue_by_country)
plt.title("Average monthly revenue by each country")
plt.xlabel("Country")
plt.ylabel("Monthly Revenue")
plt.xticks(rotation=90)
plt.show()
```

```python
num_colms=["Monthly Revenue","Age"]
sns.pairplot(data[num_colms],diag_kind="kde")
plt.suptitle("Pairplot of monthly revenue and age",y=1.02)
plt.show()
```

```python
sns.countplot(x="Subscription Type",data=data)
plt.title("Subscription Type Distibution")
plt.xlabel("Subscription Type")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
sns.histplot(data["Monthly Revenue"],bins=10,kde=True)
plt.title("Monthly revenue distribution")
plt.ylabel("Count")
plt.xlabel("Monthly Revenue")
plt.show()
```

```python
sns.histplot(data["Monthly Revenue"],bins=50,kde=True)
plt.title("Monthly revenue distribution")
plt.ylabel("Count")
plt.xlabel("Monthly Revenue")
plt.show()#bins=50
```

```python
sns.histplot(data["Monthly Revenue"],bins=5,kde=True) #bins=5
plt.title("Monthly revenue distribution")
plt.ylabel("Count")
plt.xlabel("Monthly Revenue")
plt.show()
```

```python
sns.countplot(x="Device",data=data)
plt.title("Device distribution")
plt.xlabel("Device")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```
